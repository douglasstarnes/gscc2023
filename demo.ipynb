```python
playoffs_weeks = {
    "WildCard": 18,
    "Division": 19,
    "ConfChamp": 20,
    "SuperBowl": 21
}
```

```python
games_df["week"] = games_df["week"].apply(lambda x: playoffs_weeks[x] if x in playoffs_weeks else int(x))
```

```python
titans_games = games_df[(games_df["home_team"] == "Tennessee Titans") | (games_df["away_team"] == "Tennessee Titans")]
titans_games_2017 = titans_games[titans_games["year"] == 2017]
titans_games_2017.sort_values("week")
```
