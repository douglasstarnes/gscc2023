```python
import os
```

```python
for _ in range(10):
    if fake.pyint(0, 1) == 0:
        os.mkdir(fake.word())
    else:
        f = open(f"{fake.word()}.txt", "w")
        f.write(" ".join(fake.words(10)))
        f.close()
```

```python
from collections import Counter

counts = Counter()


for item in os.listdir("."):
    if os.path.isdir(item):
        counts["directory"] += 1
    else:
        counts["file"] += 1


counts.most_common()
```
